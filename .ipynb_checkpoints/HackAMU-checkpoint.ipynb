# Now Lets Prepare our ML Model

So here we will be using scikit Learn Library to build our ML model and we will implement Random forest classifier

### Slicing our data for independent and Dependent variables

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

```python
Y
```

Here X contains an array of independent varaiables

Y contains an array of dependent variable.

### Spliting Data for Training & Testing Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

This contains  80% training dataset and 20 % testing dataset.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=300)
```

```python
model.fit(X_train,Y_train)
```

```python
pred = model.predict(X_test)
```
